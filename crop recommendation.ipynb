```python
x=crop.iloc[:,:-1]
y=crop["label"]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
label=['apple','banana','blackgram','chickpea','coconut','coffee','cotton','grapes','jute','kidneybeans','lentil','maize','mango',
      'mothbeans','mungbean','muskmelon','orange','papaya','pigeonpeas','pomegranate','rice','watermelon']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
rfr=DecisionTreeRegressor()
rfr.fit(x_train,y_train)
```

```python
rfr.score(x_test,y_test)*100, rfr.score(x_train,y_train)*100
```

```python
def recommendation(N,P,K,temperature,humidity,ph,rainfall):
    pr=rfr.predict([[N,P,K,temperature,humidity,ph,rainfall]])
    crop_name=label[int(pr)]
    return crop_name
```

```python
N=int(input("Nitrogen:"))
P=int(input("Phosphorus:"))
K=int(input("Potassium:"))
temperature= int(input("Temperature:"))
humidity=int(input("Humidity:"))
ph=int(input("Ph of Soil:"))
rainfall=int(input("Rainfall:"))
predict = recommendation(N,P,K,temperature,humidity,ph,rainfall)
print("you Should Grow",predict)
```
